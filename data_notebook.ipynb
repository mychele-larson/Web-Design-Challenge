```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import time
from scipy.stats import linregress
import seaborn as sns
```

```python
csv = "Resources/cities.csv"
```

```python
data_cities = pd.read_csv(csv)
data_cities.head()
```

```python
html = data_cities.to_html()
text_file = open("Resources/data_cities.html", "w")
text_file.write(html)
text_file.close()
plt.scatter(data_cities['Lat'],
            data_cities['Max Temp'],
            alpha=0.75,
            edgecolors='orange', linewidths=1, marker="o",)
plt.title('City Latitude vs. Max Temperature (%s)' % time.strftime('%x'))
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (C)')
plt.grid(True)
sns.set()

plt.savefig("assets/images/Lat_Temp.png")

plt.show()
```

```python

plt.scatter(data_cities['Lat'],
            data_cities['Humidity'],
            alpha=0.75,
            edgecolors='blue', linewidths=1, marker="o",)

plt.title('City Latitude vs. Humidity (%s)' % time.strftime('%x'))
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.grid(True)
sns.set()

plt.savefig("assets/images/Lat_Humidity.png")

plt.show()
```

```python
plt.scatter(data_cities['Lat'], 
            data_cities['Cloudiness'], 
            alpha = 0.75, 
            edgecolors = 'red', linewidths=1, marker="o",)

plt.title('City Latitude vs. Cloudiness (%s)' % time.strftime('%x'))
plt.xlabel('Latitude') 
plt.ylabel('Cloudiness')
plt.grid(True)
sns.set()

plt.savefig("assets/images/Lat_Clouds.png")

plt.show()
```
